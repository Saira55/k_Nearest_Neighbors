```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Define the different numbers of neighbors to test
neighbors_list = [1, 3, 5, 10, 20, 50]

# Loop over the different values of neighbors
for n_neighbors in neighbors_list:
    # Create and fit the kNN model
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(X_train, Y_train)

    # Make predictions
    Y_pred = knn.predict(X_test)

    # Ensure Y_test is one-dimensional
    # Convert DataFrame to Series (if it's a DataFrame with a single column)
    Y_test_series = Y_test.iloc[:, 0]  # Or Y_test[Y_test.columns[0]]
    # Flatten the Series to a NumPy array
    Y_test_flat = Y_test_series.values.flatten()

    # Calculate accuracy score
    accuracy = accuracy_score(Y_test_flat, Y_pred)
    print(f"Accuracy Score with {n_neighbors} neighbors:", accuracy)

    # Count correct and incorrect predictions
    correct_predictions = np.count_nonzero(Y_pred == Y_test_flat)
    incorrect_predictions = len(Y_pred) - correct_predictions
    print(f"Number of Correct Predictions with {n_neighbors} neighbors:", correct_predictions)
    print(f"Number of Incorrect Predictions with {n_neighbors} neighbors:", incorrect_predictions)
    print("-----------------------------------------")
```
